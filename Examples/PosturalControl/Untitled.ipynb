```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
tendonForce = np.linspace(0, 30, 100)
FGolgi = 68*60*np.log(tendonForce/20 + 1)
FGolgi2 = 200 * np.log(np.exp(tendonForce - 1)+1)
```

```python
plt.figure()
plt.plot(tendonForce, FGolgi/120)
plt.plot(tendonForce, FGolgi2/120)
plt.show()
```

```python
tendonLengthNorm = np.linspace(0.5, 2, 100)
Ft = 0.005 * 27.8 * np.log(np.exp((tendonLengthNorm - 0.96) / 0.005) + 1.0)
```

```python
plt.figure()
plt.plot(Ft, tendonLengthNorm)
plt.show()
```

```python
plt.figure()
plt.plot(tendonLengthNorm, FGolgi/200)
plt.plot(tendonLengthNorm, (tendonLengthNorm-1)*50)
plt.grid()
plt.show()
```
